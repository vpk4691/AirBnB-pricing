```python
from datetime import datetime, timedelta

start_date = datetime(2023, 7, 1)
end_date = datetime(2023, 9, 30)

chk_in_chk_out = {}
current_date = start_date
while current_date < end_date:
    next_date = current_date + timedelta(days=1)
    chk_in_chk_out[current_date.strftime('%Y-%m-%d')] = next_date.strftime('%Y-%m-%d')
    current_date = next_date

data = []

for listing_id in listing_ids:
    for check_in, check_out in chk_in_chk_out.items():
        url = str1 + str(listing_id) + str2 + str(check_in) + str3 + str(check_out) + str4
        data.append([listing_id, check_in, check_out, url])

df = pd.DataFrame(data, columns=["Listing ID", "Check-in", "Check-out", "URL"])
```
